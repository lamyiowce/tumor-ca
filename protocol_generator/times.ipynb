```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
times = open("protocol_list.csv")
table = []

times_dict = {}

for line in times:
    contents = line.strip().split(' ')
    table.append(contents[:-1])
    
count = 0
for line in table:
    for time in line:
        count = count + 1
        time = int(time) - (int(time) - int(round(int(time) / 1000, 0))*1000)
#         print (time)
        if time in times_dict:
            times_dict[time] = times_dict[time] + 1
        else:
            times_dict[time] = 1
    
print (count)   
```

```python
plt.bar(times_dict.keys(), times_dict.values(), 20, color='g')
```

```python
# Custom distribution of times (sampling from predefined vectors - denses at the beginning/end of intervals)
lists = sorted(times_dict.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y)
plt.show()
```

```python
# Distribution of doses

times = open("protocol_times.csv")
table = []

doses_dict = {}
doses_sum_dict = {}

for line in times:
    contents = line.strip().split(' ')
    table.append(contents[:-1])
    
    doses_sum = sum([float(i) for i in contents])
    if doses_sum in doses_sum_dict:
        doses_sum_dict[doses_sum] = doses_sum_dict[doses_sum] + 1
    else:
        doses_sum_dict[doses_sum] = 1
    
count = 0
for line in table:
    for dose in line:
        count = count + 1
#         print (time)
        if dose in doses_dict:
            doses_dict[dose] = doses_dict[dose] + 1
        else:
            doses_dict[dose] = 1
    
print (count)   

```

```python
# Custom distribution of doses

lists = sorted(doses_dict.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.figure(figsize=(30, 10))
plt.scatter([100 * xx for xx in x], y)
plt.show()
```

```python
# Custom distribution of doses sums - expected: uniform

lists = sorted(doses_sum_dict.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.show()
```
